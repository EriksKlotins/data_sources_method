```python


# cs - criteria evaluation scores
# ds - datasource evaluation scores
# im_results = False - return intermediate results
def rank_data_sources(cs, ds, im_results = False):

    
    result = {'csn': [], # criteria scores normalized
              'dsn': [], # data source scores normalized
              'ans': [], # resulting scores for each analyst
              'avs': []} # averaged scores

    # Normalize criteria scores
    result['csn'] = cs / np.sum(cs, axis=0)

    # for each scores matrix (i.e. Analyst)
    for i, scores in enumerate(ds):

        # Normalize data sources scores
        dsn = scores / np.sum(scores, axis=0)
        result['dsn'].append(dsn)

        # Dot-product between normalized datasource scores 
        # and normalized criteria scores         
        result['ans'].append(dsn @ [row[i] for row in result['csn']])


    # number of analysts        
    n = len(result['ans'])
    
    # Averaged scores
    avs = np.sum(result['ans'], axis=0) / n
    
    # scale     
    result['avs'] = avs / np.sum(avs)
    
    return result if im_results else result['avs']
    

    
```

```python
# Test cases for the method
# both should be true
    
r1 = rank_data_sources(aww, ayy, False)
t1 = [0.12888548, 0.0319326 , 0.07222175, 0.1945414 , 0.0287113, 0.04901196, 0.10743357, 0.06605435, 0.03281515, 0.04893562,0.1086552 , 0.13080162]
print(np.sum(r1 - t1)< 1e-6)


r2 = rank_data_sources(bww, byy, False)
t2 =  [0.14801177, 0.1007801 , 0.13664887, 0.15924346, 0.10225804, 0.16949001, 0.18356776]
print(np.sum(r2 - t2) < 1e-6)
```
